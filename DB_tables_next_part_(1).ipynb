```python
# Identity table
dfs_info = []
for i in range(n):
    data = pd.read_excel(files[i])
    part1 = data.iloc[:,:2]
    part2 = data.iloc[:,2:]
    part2.columns = part1.columns
    df = pd.concat((part1,part2), axis = 0, ignore_index = True)
    df1 = df.T.reset_index(drop = True)
    df1.columns = df1.iloc[0]
    df1 = df1.iloc[1:,:]
    dfs_info.append(df1)
info_df = pd.concat(dfs_info, ignore_index = True)
info_df
```

```python
# Bank Statement table
dfs_bank_info = []
for i in range(n):
    df = pd.read_excel(files[i], sheet_name = 'Bank-statement')
    name = dfs_info[i]['Name'].item()
    address = dfs_info[i]['Address'].item()
    gender = dfs_info[i]['Gender'].item()
    df['Name'] = name
    df['Address'] =  address
    df['Gender'] = gender
    dfs_bank_info.append(df)
bank_df = pd.concat(dfs_bank_info, ignore_index = True)
bank_df
```

```python
# Lifestyle table
dfs_lifestyle = []
for i in range(n):
    df = pd.read_excel(files[i], sheet_name = 'Lifestyle')
    df = df.T.reset_index(drop = True)
    df.columns = df.iloc[0]
    df = df.iloc[1:,:]
    name = dfs_info[i]['Name'].item()
    address = dfs_info[i]['Address'].item()
    gender = dfs_info[i]['Gender'].item()
    df['Gender'] = gender
    df['Name'] = name
    df['Address'] =  address
    dfs_lifestyle.append(df)
lifestyle_df = pd.concat(dfs_lifestyle, ignore_index = True)
lifestyle_df
```

```python
# Info about dependent table
dfs_dependents = []
for i in range(n):
    data = pd.read_excel(files[i], sheet_name = 'Dependents', header = None)
    part1 = data.iloc[:,:2]
    part2 = data.iloc[:,2:]
    part2.columns = part1.columns
    df = pd.concat((part1, part2), axis = 0, ignore_index = True)
    df = df.dropna(how = 'all')
    df = df.T.reset_index(drop = True)
    df.columns = df.iloc[0]
    df = df.iloc[1:,:]
    df
    def func_comb(x):
        return ", ".join(x)
    x = df['Children'].item()
    if x!=0:
        df['Age of each child'] = df.loc[:, [i+1 for i in range(x)]].astype(str).apply(func_comb, axis = 1)
    else:
        df.insert(2, 'Age of each child', np.NaN)
    df = df.drop(columns = [i+1 for i in range(x)])
    y = df[' # of family members'].item()
    if y!=0:
        df['List down the members'] = df.iloc[:, 5:5+y].astype(str).apply(func_comb, axis = 1)
    col = [i for i in range(-y+1, 0)]
    df = df.drop(columns = df.iloc[:,col])
    name = dfs_info[i]['Name'].item()
    address = dfs_info[i]['Address'].item()
    gender = dfs_info[i]['Gender'].item()
    df['Gender'] = gender
    df['Name'] = name
    df['Address'] =  address
    dfs_dependents.append(df)
dependents_df = pd.concat(dfs_dependents, ignore_index = True)
dependents_df
```

```python
# Info about guarantor table
dfs_guarantor = []
for i in range(n):
    data = pd.read_excel(files[i], sheet_name = 'Guarantor', header = None)
    part1 = data.iloc[:,:2]
    part2 = data.iloc[:,2:]
    part2.columns = part1.columns
    df = pd.concat((part1,part2), axis = 0, ignore_index = True)
    df1 = df.T.reset_index(drop = True)
    df1.columns = df1.iloc[0]
    df1 = df1.iloc[1:,:]
    dfs_guarantor.append(df1)
guarantor_df = pd.concat(dfs_guarantor, ignore_index = True)
guarantor_df.dropna(axis = 1, inplace = True)
guarantor_df
```

```python
def update_mysql_table(df, table_name):
    existing_data = pd.read_sql_table(table_name, engine)
    new_rows = df[~df.isin(existing_data.to_dict('list')).all(1)]
    if not new_rows.empty:
        new_rows.to_sql(table_name, engine, if_exists='append', index=False)
```

```python
borrower_table = 'info'
bank_table = 'bankstatements'
lifestyle_table = 'lifestyle'
dependents_table = 'dependents'
guarantor_table = 'guarantor'
update_mysql_table(info_df, borrower_table)
update_mysql_table(bank_df, bank_table)
update_mysql_table(lifestyle_df, bank_table)
update_mysql_table(dependents_df, bank_table)
update_mysql_table(guarantor_df, bank_table)
```
